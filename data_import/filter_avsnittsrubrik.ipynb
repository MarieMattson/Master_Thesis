```python
import json

# Load JSON file
with open("/mnt/c/Users/User/thesis/data_import/filtered_riksdag.json", "r", encoding="utf-8") as file:
    data = json.load(file) 

# Extract unique 'avsnittsrubrik' values
rubrik_set = {entry["avsnittsrubrik"] for entry in data if entry.get("avsnittsrubrik")}

print(rubrik_set)

def find_rubrik(keyword):
    result = {rubrik for rubrik in rubrik_set if keyword.lower() in rubrik.lower()}
    return result

# Test searches
print(find_rubrik("migration"))
print(find_rubrik("betyg"))
print(find_rubrik("inkomstskatt"))
print(find_rubrik("bistånd"))
print(find_rubrik("vindkraft"))

doc_list = ["H909103", "H90968", "H90982", "H909133", "H90981"]
```
